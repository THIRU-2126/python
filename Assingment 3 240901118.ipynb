```python
import pandas as pd
import matplotlib.pyplot as plt

try:
    # Load the data
    # Ensure the file path is correct for your local machine
    df = pd.read_csv("C:/Users/DELL/Downloads/flight_data.csv")

    # --- Task a: Display summary statistics ---
    # Using the exact column name 'Departure_Delay_Min'
    stats = df['Departure_Delay_Min'].describe()
    print("Summary Statistics for Departure Delays:")
    print(stats)
    print("\n")

    # --- Task b: Average delay per airline ---
    avg_delay = df.groupby('Airline')['Departure_Delay_Min'].mean()
    print("Average Delay per Airline:")
    print(avg_delay)
    print("\n")

    # --- Task c: Most common origin and destination ---
    common_origin = df['Origin'].mode()[0]
    common_dest = df['Destination'].mode()[0]
    print(f"Most Common Origin: {common_origin}")
    print(f"Most Common Destination: {common_dest}")
    print("\n")

    # --- Task d: Visualization ---
    plt.figure(figsize=(8, 5))
    # Using the standard plt.boxplot for better control
    plt.boxplot(df['Departure_Delay_Min'], vert=False, patch_artist=True)
    
    plt.title('Distribution of Departure Delays')
    plt.xlabel('Minutes')
    plt.grid(axis='x', linestyle='--', alpha=0.7)
    
    plt.tight_layout()
    plt.show()

except FileNotFoundError:
    print("Error: The file 'flight_data.csv' was not found at the specified path.")
except KeyError as e:
    print(f"Error: Could not find the column {e}. Check your CSV headers!")
    
```
